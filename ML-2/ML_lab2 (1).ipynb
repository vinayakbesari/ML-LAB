```python
#first convert the values into numerical
df = df.replace({'yes':1 , 'no':0 })
#ordinal
furnish_map = {'unfurnished':0 , 'semi-furnished':1 , 'furnished':2}
df = df.replace({'furnishingstatus':furnish_map})
```

```python
df.sample()
```

```python
#normalization
#1 Min-max scaler
# Manual min-max normalization for 'price' column
min_price = df['price'].min()
max_price = df['price'].max()
df['price'] = (df['price'] - min_price) / (max_price - min_price)
min_area = df['area'].min()
max_area = df['area'].max()
df['area'] = (df['area'] - min_area) / (max_area - min_area)
```

```python
df.head()
```

```python
outlier_counts = {}
for col in df.columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    low_whisk = q1 - 1.5 * iqr
    high_whisk = q3 + 1.5 * iqr
    outliers = df[(df[col] < low_whisk) | (df[col] > high_whisk)]
    outlier_counts[col] = len(outliers)

print(outlier_counts)
#price is the output
```

```python
# For only numeric columns
corr_with_price = df.corr(numeric_only=True)['price'].sort_values(ascending=False)
print(corr_with_price)
```

```python
#remove the outliers for more relevant columns , which interfere with mse,corr
# we dont remove the outliers of less relevant cols cause because they don’t strongly influence the target, and removing them would mean losing
# rows that are otherwise valuable for the important features.
#we remove the col outliers where corr > 0.4
high_corr_cols = ['area', 'bathrooms', 'airconditioning', 'stories']

mask = pd.Series(False, index=df.index)

for col in high_corr_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    low_whisk = q1 - 1.5 * iqr
    high_whisk = q3 + 1.5 * iqr
    mask |= (df[col] < low_whisk) | (df[col] > high_whisk)

df_clean = df[~mask]

print(f"Original rows: {df.shape[0]}")
print(f"Rows after cleaning: {df_clean.shape[0]}")
```

```python
Y = df_clean.iloc[:, 0]        # Select the first column as target (Price)
X = df_clean.iloc[:, 1:]       # Select the rest as features
```

```python
X.shape,Y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train_np = np.array(X_train, dtype=float)
y_train_np = np.array(y_train, dtype=float).reshape(-1, 1)
X_test_np = np.array(X_test, dtype=float)
y_test_np = np.array(y_test, dtype=float).reshape(-1, 1) #converting (n) into (nx1) matrix

# Add bias term (column of ones) to X_train and X_test
X_train_bias = np.c_[np.ones((X_train_np.shape[0], 1)), X_train_np]
X_test_bias = np.c_[np.ones((X_test_np.shape[0], 1)), X_test_np]

# θ = (Xᵀ X)⁻¹ Xᵀ y
theta = np.linalg.inv(X_train_bias.T.dot(X_train_bias)).dot(X_train_bias.T).dot(y_train_np)

# Predictions
y_pred_train = X_train_bias.dot(theta)
y_pred_test = X_test_bias.dot(theta)

# R² Score Function
def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

# Evaluation
train_r2 = r2_score(y_train_np, y_pred_train)
test_r2 = r2_score(y_test_np, y_pred_test)

print("Model coefficients (theta):", theta.flatten())
print(f"Train R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
#model is not overfitting as  r2_train >> r2_test.
#model performs equally well (or better) on unseen data.
#For predictive models, train R² must be close to test R² to ensure it generalizes well.
```

```python
# Make sure we know the feature names in the same order as X_train
feature_names = X_train.columns

# Intercept (bias term)
intercept = theta[0][0]

# Coefficients for each feature
coefficients = theta[1:].flatten()

print(f"Intercept: {intercept:.4f}")
print("\nFeature Coefficients:")
for name, coef in zip(feature_names, coefficients):
    print(f"{name:20s} {coef:.4f}")

# R² scores
print(f"\nTrain R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
```

```python
df_av['EVENT_LCL_TIME'] = df_av['EVENT_LCL_TIME'].str.replace('Z', '', regex=False)
#remove the Z
# Combine to Datetime
df_av['EVENT_DATETIME'] = pd.to_datetime(df_av['EVENT_LCL_DATE'] + ' ' + df_av['EVENT_LCL_TIME'],errors='coerce', format='%d-%b-%y %H:%M:%S')
```

```python
def extract_phase(text):
    if pd.isnull(text):
        return 'UNKNOWN'
    phases = ['LANDING', 'TAKEOFF', 'CRUISE', 'APPROACH']
    for phase in phases:
        if phase in text.upper():
            return phase
    return 'UNKNOWN'

new_df['FLIGHT_PHASE_TEXT'] = new_df['RMK_TEXT'].apply(extract_phase)
```
