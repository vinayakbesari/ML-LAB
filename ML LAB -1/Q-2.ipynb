```python
y=df["price"]
y=np.array(y)
X=df.drop(columns="price")
X=np.array(X)
```

```python
X
```

```python
#split and standardization
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
# x_train.()
```

```python
X_test=scaler.transform(X_test)
```

```python
def gradient_descent(X, y, epochs=1000, learning_rate=0.01):
    n_samples, n_features = X.shape
    
    # Initialize weights (including bias as the last term)
    w = np.zeros(n_features + 1)
    
    # Add bias column (of 1s) to the input features
    X_bias = np.hstack((X, np.ones((n_samples, 1))))

    for _ in range(epochs):
        y_pred = np.dot(X_bias, w)
        error = y - y_pred

        # Compute gradient
        gradient = -2 * np.dot(X_bias.T, error) / n_samples

        # Update weights
        w -= learning_rate * gradient

    return w
```

```python
w_gd = gradient_descent(X_train, Y_train, epochs=1000, learning_rate=0.01)

# Predict
X_test_bias = np.hstack((X_test, np.ones((X_test.shape[0], 1))))
Y_pred_gd = np.dot(X_test_bias, w_gd)

# Evaluate
mse_gd = np.mean((Y_test - Y_pred_gd) ** 2)
print("Gradient Descent MSE:", mse_gd)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(Y_test, Y_pred_gd, alpha=0.6, color='blue')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--', label='Ideal Fit')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices (Gradient Descent)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
